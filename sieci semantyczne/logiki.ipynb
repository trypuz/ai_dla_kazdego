# Zredukowane tablice analityczne

-   Tablice analityczne konstruujemy tak długo, aż uzyskamy zredukowaną
    ich postać.

-   Jeżeli $\phi$ jest wyrażeniem $\mathbb{KRZ}$, tj. gdy należy do
    $\mathbb{PZW_{KRZ}}$, to symbol $T_{\phi}$ oznacza drzewo atomowe,
    które ma w korzeniu $\phi$.

    **Definicja 3**. *Niech $T$ będzie tablicą analityczną.*

    1.  *wyrażenie $\phi$ jest *zredukowane* w $T$ jeżeli drzewo
        $T'_{\phi}$ jest poddrzewem $T$.*

    2.  *Tablica analityczna $T$ jest *zredukowana* jeżeli wszystkie
        wyrażenia z $T$ są zredukowane w $T$.*

# Zamknięte i otwarte tablice analityczne

Po uzyskaniu tablicy zredukowanej sprawdzamy, czy uzyskana tablica jest
otwarta, czy zamknięta.

**Definicja 4**. *Tablice analityczna jest *zamknięta*, jeżeli każda jej
gałąź zawiera dwa wyrażenia sprzeczne: $\phi$, $\neg \phi$.  
Tablice analityczna jest *otwarta*, jeżeli nie jest zamknięta.*

Wyrażenie $\phi \in \ensuremath{\mathbb{PZW_{KRZ}}}$ jest prawem logiki
wtedy i tylko wtedy, gdy istnieje zamknięta tablica analityczna dla
$\neg \phi$.

# Metoda tablic analitycznych - krok po kroku

Czy $\phi \in \ensuremath{\mathbb{PZW_{KRZ}}}$ jest prawem logiki?

1.  Wstaw $\neg \phi$ jako korzeń tablicy analitycznej $T$!

2.  Korzystając z atomowych tablic analitycznych zredukuj $T$!

3.  Czy $T$ jest zamknięta?

    -   <span style="color: 1.00,0.00,0.00">NIE</span> $\Longrightarrow$
        $\phi$ <span style="color: 0.98,0.00,0.00">nie jest</span>
        prawem logiki.

## Metalogiczne własności $\mathbb{KRZ}$

# Metalogiczne własności $\mathbb{KRZ}$

1.  $\mathbb{KRZ}$ jest niesprzeczny

2.  $\mathbb{KRZ}$ jest (zu)pełny

3.  $\mathbb{KRZ}$ jest rozstrzygalny

# Tezy i tautologie w logice

-   *Teza* jest to wyrażenie zdaniowe, dla którego istnieje dowód
    ($\vdash \phi$).

-   *Tautologia* jest to (zawsze) prawdziwe wyrażenie zdaniowe
    ($\vDash \phi$).

# Niesprzeczność teorii logicznej

-   *Teoria logiczna* jest to zbiór tez ("zamknięty" ze względu na
    dowody.)

-   Teoria logiczna jest *niesprzeczna* jeżeli wśród jej tez nie ma
    wyrażeń sprzecznych: $\phi$ i $\neg \phi$.

# Zdrowie i pełność teorii logicznej

-   Teoria logiczna jest zdrowa (*sound*), gdy każda jej teza jest
    tautologią.

-   Teoria logiczna jest pełna (*complete*), gdy każda jej tautologia
    jest tezą.

# Rozstrzygalność teorii logicznej

-   Teoria logiczna jest *rozstrzygalna*, gdy istnieje metoda, która w
    skończonej liczbie ściśle określonych kroków jest w stanie
    stwierdzić o każdym wyrażeniu, czy jest ono tezą tej teorii czy nie.

-   Teoria logiczna jest rozstrzygalna, gdy zbiór jej tez może zostać
    obliczony przez jakąś maszynę Turinga.

# Alfabet $\mathbb{WRP}$

**Definicja 5**. *Alfabet $\mathbb{WRP}$ zawiera:*

1.  *nieskończony zbiór zmiennych indywidualnych:
    $x, y, z, \dots, x_{1}, y_{1}, z_{1}, \dots, x_{2}, \dots$,*

2.  *nieskończony zbiór stałych indywidualnych:
    $\textrm{a}, \textrm{b}, \textrm{c}, \dots, \textrm{a}_{1}, \textrm{b}_{1}, \textrm{c}_{1}, \dots, \textrm{a}_{2}, \dots$,*

3.  *nieskończony zbiór predykatów:*

    -   *jednoargumentowych: $A, B, C, \dots, A_{1}, \dots$,*

    -   *wieloargumentowych: $P, Q, R, \dots, P_{1}, \dots$,*

4.  *zbiór funktorów prawdziwościowych:
    $\neg, \land, \lor, \to, \equiv$,*

5.  *zbiór kwantyfikatorów: $\forall, \exists$,*

6.  *nawiasy: (, ).*

# Język $\mathbb{WRP}$

Sumę zbiorów:

1.  zmiennych indywidualnych

2.  stałych indywidualnych

będziemy nazywać zbiorem *wyrażeń nazwowych* $\mathbb{WRP}$.  
Dowolny ciąg elementów alfabetu $\mathbb{WRP}$ jest *napisem*
($\mathbb{WRP}$).

# Język $\mathbb{WRP}$(2)

**Definicja 6**. *Zbiór poprawnie zbudowanych wyrażeń
$\mathbb{WRP}$ (skrót: $\mathbb{PZW_{WRP}}$) jest najmniejszym zbiorem
napisów spełniającym poniższe warunki:*

1.  *Jeżeli $\delta^{(n)}$ jest predykatem $n$-argumentowym
    $(n \geq 1)$, a $\alpha_{1}, \alpha_{2}, \dots, \alpha_{n}$ są
    wyrażeniami nazwowymi, to
    $\delta^{(n)}(\alpha_{1}, \alpha_{2}, \dots, \alpha_{n}) \in \ensuremath{\mathbb{PZW_{WRP}}}$.*

2.  *Jeśli $\phi \in \ensuremath{\mathbb{PZW_{WRP}}}$, to
    $\neg \phi \in \ensuremath{\mathbb{PZW_{WRP}}}$.*

3.  *Jeśli $\phi, \psi \in \ensuremath{\mathbb{PZW_{WRP}}}$, to
    $(\phi \land \psi), (\phi \lor \psi), (\phi \to \psi), (\phi \equiv \psi) \in \ensuremath{\mathbb{PZW_{WRP}}}$.*

4.  *Jeżeli $\phi \in \ensuremath{\mathbb{PZW_{WRP}}}$ i $\alpha$ jest
    zmienną indywidualną, to $\forall \alpha (\phi)$ i
    $\exists \alpha (\phi) \in \ensuremath{\mathbb{PZW_{WRP}}}$.*

# Zasięg kwantyfikatorów

**Definicja 7**. *W wyrażeniach $\forall \alpha~\phi$ i

# Wolność i niewola zmiennych

**Definicja 8**. *Zmienna $\alpha$ występuje w danym miejscu w wyrażeniu
$\phi$ *jako zmienna wolna*, gdy występuje w tym miejscu, lecz nie
bezpośrednio za kwantyfikatorem ani w zasięgu kwantyfikatora. Zmienna
$\alpha$ *jest wolna* w $\phi$ jeżeli występuje w $\phi$ w jakimś
miejscu jako zmienna wolna.*

**Definicja 9**. *W wyrażeniach $\forall \alpha~\phi$ i
$\exists \alpha~\phi$ kwantyfikator *wiąże* zmienną $\alpha$. Jeżeli
$\alpha$ nie jest wolne w $\phi$, to kwantyfikator wiąże $\alpha$
*nieistotnie*.*

## Semantyka dla $\mathbb{WRP}$

# Modele $\mathbb{WRP}$

**Definicja 12**. *Parę $<\mathbb{D}, \mathbb{I}>$ nazywamy modelem dla
języka $\mathbb{WRP}$  jeżeli:*

1.  *$\mathbb{D} \neq \emptyset$,*

2.  *$\mathbb{I}$ jest taką funkcją, która:*

    1.  *odwzorowuje zbiór $\mathbb{CONST}$ stałych języka
        $\mathbb{WRP}$ w zbiór $\mathbb{D}$,*

    2.  *odwzorowuje zbiór predykatów **jedno**argumentowych
        $\mathbb{REL}$ w $\wp(\mathbb{D})$ (=zbiór wszystkich podzbiorów
        $\mathbb{D}$),*

    3.  *odwzorowuje zbiór predykatów **dwu**argumentowych
        $\mathbb{REL}$ w $\wp(\mathbb{D} \times \mathbb{D})$ (=zbiór
        wszystkich podzbiorów iloczynu kartezjańskiego $\mathbb{D}$
        “razy” $\mathbb{D}$)*

    4.  *…,*

    5.  *odwzorowuje zbiór predykatów $n$-argumentowych $\mathbb{REL}$ w
        $\wp(\mathbb{D}^{n})$ (=zbiór wszystkich podzbiorów
        $n$-argumentowego iloczynu kartezjańskiego
        $\underbrace {\mathbb{D} \times \mathbb{D} \times \dots \times \mathbb{D}}_{n \mathrm{~razy}}$).*

*Zbiór $\mathbb{D}$ nazywamy dziedziną modelu.*

# Wartościowanie w $\mathbb{WRP}$

**Definicja 13**. *Niech $\mathfrak{M}=<\mathbb{D}, \mathbb{I}>$ będzie
modelem $\mathbb{WRP}$.*

-   Niech $x \in \mathbb{D}$.

-   Jeżeli $\mathfrak{v}$ jest wartościowaniem, to
    $\mathfrak{v}[\alpha/x]$ jest takim wartościowaniem, że:
    $$\mathfrak{v}[\alpha/x](\beta)= \left\{
    \begin{array}{rl}
    \mathfrak{v}(\beta) & \text{jeżeli } \alpha \ne \beta \\
    x & \text{jeżeli } \alpha = \beta.
    \end{array} \right.$$

# Spełnianie w $\mathbb{WRP}$

Niech $\mathfrak{M}=<\mathbb{D}, \mathbb{I}>$ będzie modelem
$\mathbb{WRP}$.  

-   $\mathfrak{M},\mathfrak{v} \vDash \phi$, to tyle, co $\phi$ jest
    spełnione w modelu $\mathfrak{M}$ przy wartościowaniu
    $\mathfrak{v}$.

-   $\mathfrak{M},\mathfrak{v}  \nvDash \phi$ to tyle, co $\phi$ **nie**
    jest spełnione w modelu $\mathfrak{M}$ przy wartościowaniu
    $\mathfrak{v}$.

-   $\mathfrak{M} \vDash \phi$, to tyle, co $\phi$ jest spełnione w
    modelu $\mathfrak{M}$.

-   $\vDash \phi$, to tyle, co $\phi$ jest tautologią.

# Spełnianie w $\mathbb{WRP}$ (2)

$$\begin{aligned}
\label{sat-2}
& \mathfrak{M},\mathfrak{v} \vDash \delta_n^{(s_{n})}(\alpha_{1}, \alpha_{2}, \dots, \alpha_{s_{n}}) \equiv <\mathfrak{v}(\alpha_{1}), \mathfrak{v}(\alpha_{2}), \dots, \mathfrak{v}(\alpha_{s_{n}})> \in \mathbb{I}(\delta_n^{(s_{n})})\nonumber \\ \\
%
& \mathfrak{M},\mathfrak{v} \vDash \neg \phi \equiv \mathfrak{M} \nvDash \phi \\
%
& \mathfrak{M},\mathfrak{v} \vDash \phi \land \psi \equiv \mathfrak{M} \vDash \phi \textrm{~oraz~}  \mathfrak{M} \vDash \psi\\
%
& \mathrm{itd.}\nonumber \\
%
& \mathfrak{M},\mathfrak{v} \vDash \forall \alpha ~ \phi \equiv \mathrm{~dla~ dowolnego~} x \in \mathbb{D}~\mathfrak{M},\mathfrak{v}[\alpha/x] \vDash \phi\\
%
& \mathfrak{M},\mathfrak{v} \vDash \exists \alpha ~ \phi \equiv \mathrm{~dla~ jakiegoś~} x \in \mathbb{D}~\mathfrak{M},\mathfrak{v}[\alpha/x] \vDash \phi\end{aligned}$$

# Spełnianie w $\mathbb{WRP}$ (3)

$$\mathfrak{M} \vDash \phi \equiv \forall \mathfrak{v}~\mathfrak{M},\mathfrak{v} \vDash \phi.$$
$$\vDash \phi \equiv \forall \mathfrak{M}~ \mathfrak{M}\vDash \phi.$$

# Spełnianie w $\mathbb{WRP}$ a tablice analityczne

-   Jeżeli $\vDash \phi$, to istnieje skończona zamknięta tablica
    analityczna dla $\phi$.

-   Jeżeli $\neg \vDash \phi$, to istnieje, skończona lub nieskończona,
    otwarta tablica analityczna dla $\phi$.

    subsumpcji pomiędzy pojęciami.

**Definicja 14**. *Alfabet $\mathcal{ALC}$ zawiera:*

        $$\mathbb{ROLE} := \{R_{1}, R_{2}, \dots\}$$*

3.  *funktory (nazwotwórcze od argumentów nazwowych): $-$, $\sqcap$;*

6.  *nawiasy: (, ).*

# “Kwantyfikatory” w logikach opisowych

-   $\exists R . C$ - (predykat reprezentujący) zbiór takich
    przedmiotów, które są powiązane relacją $R$ z *przynajmniej jednym*
    przedmiotem ze zbioru $C$
    $$\exists R . C(x) \equiv \exists y [R(x, y) \land C(y)]$$

-   $\forall R . C$ - (predykat reprezentujący) zbiór takich
    przedmiotów, które są powiązane relacją $R$ *tylko* z przedmiotami
    ze zbioru $C$
    $$\forall R . C(x) \equiv \forall y [R(x, y) \to C(y)]$$

-   przykłady:

    -   $R$ - jest potomkiem, $C$ - jest Polakiem

        -   $\exists R . C$ - potomkowie (jakichś) Polaków

        -   $\forall R . C$ - prawdziwi Polacy (potomkowie tylko
            Polaków)

    -   $R$ - zna, $C$ - jest celebrytą

        -   $\exists R . C$ - ci, którzy znają jakiegoś celebrytę

        -   $\forall R . C$ - ci, którzy znają tylko celebrytów

do *przeciwdziedziny* relacji $R$ wtedy i tylko wtedy, gdy
$$\exists y~ R(y, x).$$*

# (Najważniejsze) rodzaje relacji dwuargumentowych

1.  zwrotne i przeciwzwrotne

2.  symetryczne, asymetryczne, antysymetryczne

3.  przechodnie

4.  funkcje

# Relacje zwrotne

**Definicja 21**. *Relacja dwuargumentowa $R$ jest *zwrotna* wtedy i
tylko wtedy, gdy $$\forall x ~R(x,x).$$*

przykłady:

1.  równoległość (prostych)

2.  $=$

3.  podobieństwo

antyprzykłady:

1.  prostopadłość (prostych)

2.  $<$

3.  bycie dłużnikiem

4.  nienawiść

# Relacje przeciwzwrotne

**Definicja 22**. *Relacja dwuargumentowa $R$ jest *przeciwzwrotna*
wtedy i tylko wtedy, gdy $$\forall x ~\neg R(x,x).$$*

przykłady:

1.  prostopadłość (prostych)

2.  $<$

3.  bycie dłużnikiem

antyprzykłady:

1.  równoległość (prostych)

2.  $=$

3.  podobieństwo

4.  nienawiść

# Relacje symetryczne

**Definicja 23**. *Relacja dwuargumentowa $R$ jest *symetryczna* wtedy i
tylko wtedy, gdy $$\forall x,y ~[R(x,y) \to R(y, x)].$$*

przykłady:

1.  równoległość (prostych)

2.  $=$

3.  podobieństwo

4.  prostopadłość (prostych)

antyprzykłady:

1.  $<$

2.  bycie dłużnikiem

# Relacje asymetryczne

**Definicja 24**. *Relacja dwuargumentowa $R$ jest *asymetryczna* wtedy
i tylko wtedy, gdy $$\forall x,y ~[R(x,y) \to \neg R(y, x)].$$*

przykłady:

1.  $<$

2.  bycie dłużnikiem

antyprzykłady:

1.  równoległość (prostych)

2.  prostopadłość (prostych)

3.  $=$

4.  podobieństwo

# Relacje antysymetryczne

**Definicja 25**. *Relacja dwuargumentowa $R$ jest *antysymetryczna*
wtedy i tylko wtedy, gdy
$$\forall x,y ~[R(x,y) \land R(y, x) \to x=y].$$*

przykłady:

1.  $\leq$

antyprzykłady:

1.  równoległość (prostych)

2.  prostopadłość (prostych)

3.  podobieństwo

4.  $<$

5.  bycie dłużnikiem

# Relacje przechodnie

**Definicja 26**. *Relacja dwuargumentowa $R$ jest *przechodnia* wtedy i
tylko wtedy, gdy $$\forall x,y,z ~[R(x,y) \land R(y, z) \to R(x,z)].$$*

przykłady:

1.  równoległość (prostych)

2.  $<$

3.  $=$

antyprzykłady:

1.  prostopadłość (prostych)

2.  podobieństwo

# Funkcje

**Definicja 27**. *Relacja dwuargumentowa $R$ jest *funkcją* wtedy i
tylko wtedy, gdy
$$\forall x,y_{1},y_{2} ~[R(x,y_{1}) \land R(x, y_{2}) \to y_{1}=y_{2}].$$*

przykłady:

1.  $=$

2.  $x$ urodził się w dniu $y$

3.  $x$ ma numer PESEL $y$

antyprzykłady:

1.  równoległość (prostych)

2.  prostopadłość (prostych)

3.  podobieństwo

4.  $<$

5.  bycie dłużnikiem

# Relacje (ściśle) porządkujące

**Definicja 28**. *Relacja jest ściśle porządkująca (jest porządkiem)
jeżeli jest asymetryczna i przechodnia.*

przykłady:

1.  $<$

2.  być głupszym niż

antyprzykłady:

1.  równoległość (prostych)

2.  $\leq$

# Relacje równoważnościowe

**Definicja 29**. *Relacja jest równoważnościowa jeżeli jest zwrotna,
symetryczna i przechodnia.*

przykłady

1.  równoległość (prostych)

2.  równoważenie się na wadze ($x$ waży tyle samo, co $y$)

antyprzykłady:

1.  $\leq$

2.  bycie dłużnikiem

3.  być niegłupszym niż

Dzięki tzw. twierdzeniu o abstrakcji relacje równoważnościowe mogą
służyć do definiowania własności, np.

1.  równoległość definiuje kierunek (prostej)

2.  równoważenie się na wadze definiuje ciężar

# Operacje na relacjach

-   ogólne (teoriomnogościowe): suma, iloczyn, różnica, dopełnienie,
    itd.

-   lokalne:

    1.  inwers relacji

    2.  iloczyn względny relacji

# Iloczyn względny relacji

**Definicja 30**. *Relacja dwuargumentowa $R$ jest *iloczynem względnym
relacji* $S_1$ i $S_2$ wtedy i tylko wtedy, gdy
$$\forall x,y ~[R(x,y) \equiv \exists z (S_{1}(x,z) \land S_{2}(z,y))].$$*

przykłady:

1.  bycie wujkiem jest iloczynem względnym relacji bycia bratem i bycia
    matką

2.  bycie wnukiem jest iloczynem względnym relacji bycia dzieckiem i
    bycia dzieckiem

3.  $<$ jest ilocznym względnym $<$ i $<$

# Inwers relacji

**Definicja 31**. *Relacja dwuargumentowa $R$ jest *inwersem relacji*
$S$ wtedy i tylko wtedy, gdy $$\forall x,y ~[R(x,y) \equiv S(y,x)].$$*

przykłady:

1.  bycie rodzicem jest inwersem relacji bycia dzieckiem

2.  $<$ jest inwersem relacji $>$

3.  równoległość prostych jest swoim własnym inwersem

## Logika opisowa $\mathcal{SROIQ}$

# Jaka logika opisowa jest najlepsza i dlaczego jest to $\mathcal{SROIQ}$?

-   

-   nazwy indywidualne

-   pojęcia

    -   pojęcia atomiczne (włącznie z $\top$)

    -   operacje na pojęciach

        -   (uniwersalna) negacja $\neg$ oraz iloczyn $\sqcap$

        -   suma $\sqcup$

        -   $\exists R. C$ oraz $\forall R.C$

        -   $\geq n R . C$ - (predykat reprezentujący) zbiór takich
            przedmiotów, które są powiązane relacją $R$ z *przynajmniej*
            $n$ przedmiotami ze zbioru $C$

        -   $\leq n R . C$ - (predykat reprezentujący) zbiór takich
            przedmiotów, które są powiązane relacją $R$ z *conajwyżej*
            $n$ przedmiotami ze zbioru $C$

        -   $\exists R.\texttt{Self}$ - (predykat reprezentujący)
            iloczyn roli $R$ oraz relacji identyczności

        -   operator do tworzenia tzw. nazw nominalnych (*nominals*),
            np. $\{\text{Elvis Presley}\}$

-   role

    -   rola uniwersalna $U$

    -   identyczność ($=$)

    -   inwersja ról ($^{-}$)

    -   (wieloargumentowy) iloczyn względny ról ($\circ$)

-   negacja (przy-)zdaniowa: “$\sim$”

# Jaka logika opisowa jest najlepsza i dlaczego jest to $\mathcal{SROIQ}$? (2)

-   zawieranie się i równoważność pojęć i ról

-   stwierdzenia (*assertions*) i negacje stwierdzeń

-   formalne własności ról oraz rozłączność ról

# Jak powiedzieć w $\mathcal{SROIQ}$ tego, czego nie można powiedzieć w $\mathcal{ALC}$?

-   Każdy kij ma dwa końce.
    $$Stick ~\sqsubseteq~ (\geq 2 hasEnd . \top) \sqcap (\leq 3 hasEnd . \top)$$

-   Każdy, kto zbudował dom, posadził też drzewo.
    $$\exists build . House \sqsubseteq \exists plant . Tree$$

-   Przyjaciele moich przyjaciół są moimi przyjaciółmi.
    $$hasFriend \circ hasFriend \sqsubseteq hasFriend$$

# Czego (nadal) nie można powiedzieć w $\mathcal{SROIQ}$?

-   Dla każdej liczby istnieje liczba od niej większa.

-   Istnieje przyczyna wszystkiego.

# Alfabet $\mathcal{SROIQ}$ – DEFINICJA

Alfabet $\mathcal{SROIQ}$ zawiera:

1.  zbiór stałych indywidualnych
    $\mathbb{CONST} := \{a_{1}, a_{2}, \dots\},$

2.  zbiór predykatów

    -   jednoargumentowych (pojęć atomicznych):
        $$\mathbb{CONCEPT} := \{\top, A_{1}, A_{2}, \dots\}$$

    -   dwuargumentowych (ról atomicznych):
        $$\mathbb{ROLE} := \{\texttt{U}, =, R_{1}, R_{2}, \dots\}$$

3.  funktory (nazwotwórcze od argumentów nazwowych): $\neg$, $\sqcap$,
    $\sqcup$, $\circ$, $^{-}$, $\{\dots\}$;

4.  “kwantyfikatory”: $\exists .$, $\forall .$, $\geq n .$, $\leq n .$,
    $\exists . \texttt{Self}$

5.  funktory (zdaniotwórcze od argumentów nazwowych): $\sqsubseteq$,
    $\equiv$, $\texttt{Dis}$, $=$, $\texttt{Refl}$, $\texttt{Irrefl}$,
    $\texttt{Sym}$, $\texttt{Asym}$, $\texttt{Trans}$

6.  funktor (zdaniotwórczy od argumentu zdaniowego): $\sim$;

7.  nawiasy: (, ).

# Zbiór pojęć logiki $\mathcal{SROIQ}$

**Definicja 32**. *Zbiór pojęć logiki $\mathcal{SROIQ}$ jest
najmniejszym zbiorem napisów $\mathcal{SROIQ}$ spełniającym poniższe
warunki:*

1.  *Każde pojęcie atomiczne jest pojęciem $\mathcal{SROIQ}$.*

2.  *Jeżeli $\alpha$ jest stałą nazwową, to $\{ \alpha \}$ jest
    pojęciem.*

3.  *Jeżeli $\delta$ jest pojęciem, to $\neg \delta$ jest pojęciem.*

4.  *Jeżeli $\delta_{1}, \delta_{2}$ są pojęciami, to
    $\delta_{1} \sqcap \delta_{2}$ i $\delta_{1} \sqcup \delta_{2}$ są
    pojęciami.*

5.  *Jeżeli $\rho$ jest rolą i $\delta$ jest pojęciem, to
    $\exists \rho . \delta$, $\forall \rho . \delta$ są pojęciami.*

6.  *Jeżeli $\rho$ jest prostą rolą, $\delta$ jest pojęciem i $n$ jest
    liczbą naturalną, to $\geq n \rho . \delta$ i $\leq n \rho . \delta$
    są pojęciami.*

7.  *Jeżeli $\rho$ jest prostą rolą, to $\exists \rho. \texttt{Self}$
    jest pojęciem.*

# Regularny porządek ról

**Definicja 38**. *Niech $\mathcal{R}$ będzie zbiorem ról logiki
$\mathcal{SROIQ}$. Przeciwzwrotna i przechodnia relacja $<$ w zbiorze
$\mathcal{R}$ jest *regularnym porządkiem*, gdy dla każdych
$\rho_1, \rho_2 \in \mathcal{R}$
$$\rho_{1} < \rho_{2} \equiv \rho_{1}^{-} < \rho_{2}$$*

# Regularne aksjomaty ról

**Definicja 39**. *Niech
$\rho_{1} \circ \rho_{2} \circ \dots \circ \rho_{n}$ będzie rolą logiki
$\mathcal{SROIQ}$ taką że żadna z ról
$\rho_{1}, \rho_{2}, \dots, \rho_{n}$ nie jest rolą uniwersalną.
Aksjomat subsumpcji ról
$\rho_{1} \circ \rho_{2} \circ \dots \circ \rho_{n} \sqsubseteq \rho$
($\rho \neq \texttt{U}$) jest $<$-regularny jeżeli zachodzi jeden z
następujących warunków:*

1.  *$n=1$ i $\rho_{1} = \rho^{-}$*

2.  *$n=2$ i $\rho_{1}=\rho_{2}=\rho$*

3.  *$\rho_{i} < \rho$ (dla $1 \leq i \leq n$)*

4.  *$\rho_{1}=\rho$ i $\rho_{i} < \rho$ (dla $1 < i \leq n$)*

5.  *$\rho_{n}=\rho$ i $\rho_{i} < \rho$ (dla $1 \leq i < n$)*

# Regularne zbiory aksjomatów ról

**Definicja 40**. *Hierarchia ról jest $<$-regularna jeżeli każdy
aksjomat w tej hierarchii jest $<$-regularny.*

# Role proste

Niech
$\phi=\rho_{1} \circ \rho_{2} \circ \dots \circ \rho_{n} \sqsubseteq \rho$
będzie aksjomatem subsumpcji ról i niech $\Gamma'$ i $\Gamma''$ będą
łańcuchami ról.  
Niech $\mathcal{A}$ będzie zbiorem aksjomatów subsumpcji ról.  
$\sqsubseteq_{\mathcal{A}}^{*}$ jest zwrotno-tranzytywnym domknięciem
$\sqsubseteq_{\mathcal{A}}$.

# Role proste (2)

Niech $\Gamma = \rho_{1} \circ \rho_{2} \circ \dots \circ \rho_{n}$
będzie rolą. Niech $\mathcal{A}$ będzie zbiorem aksjomatów subsumpcji
ról.

# Role proste (3)

**Definicja 41**. *Rola $\rho$ jest prosta ze względu na zbiór
$\mathcal{A}$ aksjomatów subsumpcji ról, gdy
$\rho_{1} \circ \rho_{2} \circ \dots \circ \rho_{n} \sqsubseteq_{\mathcal{A}^{-}}^{*} \rho$
implikuje, że $n=1$.*

# Role proste (2)

**Definicja 42**. *Rola $\rho$ jest *prosta* ze względu na pudełko
relacyjne $R$, gdy*

1.  *nie występuje po prawej stronie aksjomatów hierarchii ról w $R$
    lub*

2.  *jeżeli występuje po prawej stronie jakiegoś aksjomatu hierarchii
    ról, to dla każdego takiego aksjomatu rola, która występuje po lewej
    stronie, też jest prosta lub*

3.  *jest inwersem roli prostej.*

# Role proste (3)

**Definicja 43**. *Zbiór asercji ról jest prosty gdy wszystkie aksjomaty
występujące w nim o postaci $\texttt{Irrefl}(\rho)$,
$\texttt{Asym}(\rho)$ lub $\texttt{Dis}(\rho_{1}, \rho_{2})$, zawierają
tylko role proste.*
